# GENDER


```python
df.columns
```

```python
ax=sns.countplot(x='Gender', data=df)
```

```python
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    
    
```

```python

df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
gen_sales=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount', data=gen_sales)
```

we can conclude from the above graph that the most of the buyers were females

# age


```python
df.columns
```

```python
sns.countplot(data= df, x='Age Group', hue='Gender')
```

```python
ax=sns.countplot(data= df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
age_sales=df.groupby(['Age Group'] , as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y="Amount",data=age_sales)
```

from above graphs we have found that the most buyers were from the age group 26-35 and the bought more

# state


```python
df.columns
```

```python
#total number of orders form top10 states
state_sales=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(20)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=state_sales,x='State',y='Orders')
```

```python
#total amount/sales form top10 states
state_ord=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(20)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=state_ord,x='State',y='Amount')
```

we can see that most of the orders are from up, maharashtra, karnataka

# marital status

```python
ax=sns.countplot(data=df, x="Marital_Status")
sns.set(rc={'figure.figsize':(7,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# occupation


```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df, x="Occupation")
sns.set(rc={'figure.figsize':(7,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occ_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=occ_sales)
sns.set(rc={'figure.figsize':(10,5)})
```

from the above graphs most of the buyers were from IT sector,healthcare, aviation

# product category


```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df, x="Product_Category")
sns.set(rc={'figure.figsize':(7,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occ_sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=occ_sales)

```

we can conclude tha category from food, clothing,electronics and gadgets

```python
top_sales=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y='Orders',data=top_sales)
```

MARRIED WOMEN FROM AGE GROUP 26-35 YRS FROM UP,MAHARASHTRA,KARNATAKA from IT sector,healthcare, aviation ARE MORE LIKELY TO BUY ITEMS FROM CATEGORY food, clothing,electronics and gadgets.
